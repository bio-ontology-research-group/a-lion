```python
import mowl
mowl.init_jvm("10g")
```

```python
from mowl.datasets.base import PathDataset
source_owl = "../data/mouse.owl"
target_owl = "../data/human.owl"

source_ds = PathDataset(source_owl)
target_ds = PathDataset(target_owl)
```

```python
from mowl.projection.owl2vec_star.model import OWL2VecStarProjector

proj = OWL2VecStarProjector(bidirectional_taxonomy = True, only_taxonomy = False, include_literals = True)

source_graph = proj.project(source_ds.ontology)
target_graph = proj.project(target_ds.ontology)
```

```python
def format_value(value):
    value = value.lower()
    value = value.replace(" ", "_")
    return value
```

```python
new_source_graph = []
for edge in source_graph:
    if edge.rel() == "rdfs:label":
        value = format_value(edge.dst())
        new_source_graph.append(Edge(edge.src(), edge.rel(), value))
        new_source_graph.append
    else:
        new_source_graph.append(edge)
        
new_target_graph = []
for edge in target_graph:
    if edge.rel() == "rdfs:label":
        value = format_value(edge.dst())
        new_target_graph.append(Edge(edge.src(), edge.rel(), value))
    else:
        new_target_graph.append(edge)
        
    
```

```python
from mowl.projection.edge import Edge
src_classes, relations = Edge.getEntitiesAndRelations(new_source_graph)
tar_classes, _ = Edge.getEntitiesAndRelations(new_target_graph)
```
